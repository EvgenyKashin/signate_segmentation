```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('/mnt/ssd/kashin/ai_edge/segmentation/runs/full_labels/train_0.log', 'r') as f:
    data = f.readlines()
```

```python
loss_step = []
loss = []
val_loss_step = []
val_loss = []
ious = []

for d in data:
    d = json.loads(d)
    if 'loss' in d:
        loss_step.append(d['step'])
        loss.append(d['loss'])
    if 'valid_loss' in d:
        val_loss_step.append(d['step'])
        val_loss.append(d['valid_loss'])
        ious.append(d['iou'])
```

```python
len(loss_step), len(val_loss_step)
```

```python
max(ious)
```

```python
ious.index(max(ious))
```

```python
plt.plot(loss_step, loss, label='loss', alpha=0.5)
plt.plot(val_loss_step, val_loss, label='val_loss')
plt.plot(val_loss_step, ious, label='iou')
plt.axvline(val_loss_step[ious.index(max(ious))])
plt.legend()
plt.ylim(0, 1)
```
